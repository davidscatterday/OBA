```python
import streamlit as st
import sqlite3
import pandas as pd


# Connect to the SQLite database
conn = sqlite3.connect('nycprocurement.db')

# Function to get the first 5 rows from the table
def get_first_five_rows():
    query = """
    SELECT *
    FROM newtable
    LIMIT 5
    """
    df = pd.read_sql_query(query, conn)
    return df

# Get the first 5 rows
first_five_rows = get_first_five_rows()

# Display the first 5 rows
display(first_five_rows)

```

```python
# Streamlit app
st.title("NYC Procurement Database - First 5 Rows")

# Get and display the first 5 rows
first_five_rows = get_first_five_rows()

if not first_five_rows.empty:
    st.dataframe(first_five_rows)
else:
    st.write("No data found in the table.")

# Display the column names
st.subheader("Column Names:")
st.write(", ".join(first_five_rows.columns))

# Close the database connection
conn.close()
```
